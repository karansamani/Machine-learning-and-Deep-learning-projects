```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
T5_cosine_similarity_value=86.25
BERT_cosine_similarity_value=56.8
BioBERT_cosine_similarity_value=97.6



T5_GPT_value=94.36
BERT_GPT_value=70.1
BioBERT_GPT_value=99

```

```python
accuracy_list=[BERT_cosine_similarity_value,T5_cosine_similarity_value,BioBERT_cosine_similarity_value]

#accuracy_list.sort()
# print(accuracy_list)

classifier_names_list=['BERT','T5','BioBERT']
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(5, 5))

ax=sns.barplot(x=classifier_names_list, y=accuracy_list)
patches = ax.patches
for i in range(len(patches)):
   x = patches[i].get_x() + patches[i].get_width()/2
   y = patches[i].get_height()+0.8
   ax.annotate('{:.2f}%'.format(accuracy_list[i]), (x, y), ha='center')




plt.xlabel("NLP Q&A Models",y=2)
plt.ylabel("Cosine_similarity_accuracy Score",x=3)
#plt.xticks(rotation=60)
plt.title("Model Comparison - Cosine_similarity_Accuracy",y=1.1)
#plt.grid(True)
plt.show()
```
